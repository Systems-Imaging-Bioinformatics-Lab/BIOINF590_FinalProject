```python
original = im

def join_horizontally(*args):
    """Joins many PIL images of the same dimensions horizontally"""
    w, h = args[0].size
    n = len(args)
    joined = Image.new("RGB", (n*w, h))
    for x_off, img in zip(range(0, n*w, w), args):
        joined.paste(img, (x_off, 0))
    return joined

# stain_dict = {'eosin': [0.07, 0.99, 0.11], 'null': [0.0, 0.0, 0.0],
#               'hematoxylin': [0.65, 0.7, 0.29], 'dab': [0.27, 0.57, 0.78]}


# Create a deconvolution object with the image
dec = Deconvolution(image=original, basis=[[0.91, 0.38, 0.71], [0.39, 0.47, 0.85],[0.0, 0.0, 0.0]])
# dec = Deconvolution(image=original, basis=[stain_dict['hematoxylin'], stain_dict['eosin']])


# Produce density matrices for both colors. Be aware, as Beer's law do not always hold.
(first_density, second_density,third_density) = dec.out_scalars()
print(first_density.shape, second_density.shape)

# Produce reconstructed image, first layer, second layer and rest
out_images = dec.out_images(mode=[0, 1, 2, 3])
# Original image, reconstructed, layers and rest
joined_im = join_horizontally(original, *out_images)

plt.imshow(joined_im)
plt.show()

# fig2 = plt.figure(figsize=(10,15))
# ax2 = fig2.add_subplot(121)
# ax2.imshow(original)
# ax2 = fig2.add_subplot(122)
# plt.imshow(out_images[0])
# plt.show()

fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(131)
aIm = ax.imshow(first_density,cmap='binary')
plt.colorbar(aIm,ax=ax)
ax = fig.add_subplot(132)
aIm = plt.imshow(second_density,cmap='binary')
plt.colorbar(aIm,ax=ax)
ax = fig.add_subplot(133)
aIm = plt.imshow(third_density,cmap='binary')
plt.colorbar(aIm, ax=ax)
plt.show()


import deconvolution.pixeloperations as po

basisVec = dec.pixel_operations.get_basis()
pxO= dec.pixel_operations
_white255 = np.array([255, 255, 255], dtype=float)
(d_1, d_2, d_3) = dec.out_scalars()
invDim = d_1.shape
print(d_1.shape)
# first_rgb = np.tile(first_density,(1,1,3))
# print(first_rgb.shape)
iDimRGB = (invDim[0],invDim[1],3)
# rgb_1 = np.zeros(iDimRGB)
# rgb_1[:,:,0] = basisVec[0,0] ** d_1 
# rgb_1[:,:,1] = basisVec[0,1] ** d_1 
# rgb_1[:,:,2] = basisVec[0,2] ** d_1 


v, u, w = pxO.get_basis()
vf, uf, wf = np.zeros(iDimRGB), np.zeros(iDimRGB), np.zeros(iDimRGB)
vf[:], uf[:], wf[:] = v, u, w

a, b, c = map(po._array_positive, dec.out_scalars())
af = np.repeat(a, 3).reshape(iDimRGB)
af = af * .5
bf = np.repeat(b, 3).reshape(iDimRGB)
cf = np.repeat(c, 3).reshape(iDimRGB)

rgbOut = po._array_to_colour_255(_white255 * (vf ** af) * (uf ** bf) * (wf ** cf))
rgb_1 = po._array_to_colour_255(_white255 * (vf ** af))
rgb_2 = po._array_to_colour_255(_white255 * (uf ** bf))
rgb_3 = po._array_to_colour_255(_white255 * (wf ** cf))


# first_rgb = np.array(np.minimum(np.maximum(first_rgb, 0), 255), dtype=np.uint8)
print(bf.shape)


fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(151)
aIm = plt.imshow(im)
ax = fig.add_subplot(152)
aIm = plt.imshow(rgbOut)
ax = fig.add_subplot(153)
aIm = ax.imshow(rgb_1)
ax = fig.add_subplot(154)
aIm = plt.imshow(rgb_2)
ax = fig.add_subplot(155)
aIm = plt.imshow(rgb_3)
# plt.colorbar(aIm)
plt.show()
```
